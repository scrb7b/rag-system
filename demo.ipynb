```python
from src.config import settings

print(f"LLM provider : {settings.llm_provider}")
print(f"LLM model    : {settings.ollama_model if settings.llm_provider == 'ollama' else settings.openai_model}")
print(f"Embed model  : {settings.embed_model}")
print(f"Reranking    : {settings.enable_reranking}")
print(f"Top-K        : {settings.top_k}")
```

```python
from src.ingestion.loaders import load_directory

chunks = load_directory('./data')

print(f"Завантажено чанків: {len(chunks)}")
print(f"\nПриклад чанку:")
print(f"  source : {chunks[0]['metadata']['filename']}")
print(f"  text   : {chunks[0]['text'][:200]}...")
```

```python
from src.ingestion.embedder import VectorStore

store = VectorStore()
added = store.add_chunks(chunks)

print(f"Проіндексовано: {added} чанків")
print(f"Всього в базі : {store.count()}")
```

```python
query = "Яка молекулярна маса гліфосату?"
results = store.search(query, top_k=10)

print(f"Запит: {query}\n")
for i, r in enumerate(results, 1):
    score_key = 'rerank_score' if 'rerank_score' in r else 'score'
    print(f"[{i}] score={r[score_key]:.4f} | {r['metadata'].get('filename', '?')}")
    print(f"     {r['text'][:150]}")
    print()
```

```python
from src.agent import QAAgent
from src.generation import get_llm

agent = QAAgent(store, get_llm())

questions = [
    "Яка молекулярна маса гліфосату?",
    "До якого хімічного класу належить гліфосат?",
    "Де можна розміщувати кукурудзу у польових сівозмінах?",
    "Яка норма витрати гербіциду Акріс® на кукурудзі?",
]

for q in questions:
    result = agent.ask(q)
    print(f"Q: {q}")
    print(f"A: {result.answer}")
    print(f"   Джерела: {', '.join(result.sources)}")
    print()
```

```python
while True:
    try:
        question = input("Ваше питання (або 'exit'): ").strip()
        if not question or question.lower() in ('exit', 'q', 'quit'):
            break
        result = agent.ask(question)
        print(f"\nВідповідь: {result.answer}")
        print(f"Джерела  : {', '.join(result.sources)}\n")
    except KeyboardInterrupt:
        break
```
